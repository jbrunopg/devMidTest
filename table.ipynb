```python
import tkinter as tk

def multiplication_table():
    try:
        number = int(entry.get())
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, f"Multiplication table of {number}:\n")
        for i in range(1, 11):
            result = number * i
            result_text.insert(tk.END, f"{number} x {i} = {result}\n")
    except ValueError:
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, "Invalid input. Please enter a valid integer.")

# Create the main window
root = tk.Tk()
root.title("Multiplication Table")

# Create widgets
label = tk.Label(root, text="Enter a number:")
label.pack(pady=10)

entry = tk.Entry(root)
entry.pack(pady=5)

button = tk.Button(root, text="Show Table", command=multiplication_table)
button.pack(pady=10)

result_text = tk.Text(root, width=30, height=15)
result_text.pack()

# Run the Tkinter event loop
root.mainloop()
```
