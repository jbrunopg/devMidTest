```python
import tkinter as tk

def is_palindrome(word):
    word = word.lower()
    return word == word[::-1]

def check_palindrome():
    user_input = entry.get()
    result = is_palindrome(user_input)
    if result:
        result_label.config(text=f"{user_input} é um palíndromo!")
    else:
        result_label.config(text=f"{user_input} não é um palíndromo!")

# Configuração básica da janela
window = tk.Tk()
window.title("Verificar Palíndromo")
window.geometry("300x200")

# Elementos da interface
entry_label = tk.Label(window, text="Digite uma palavra:")
entry_label.pack(pady=10)
entry = tk.Entry(window)
entry.pack()
check_button = tk.Button(window, text="Verificar Palíndromo", command=check_palindrome)
check_button.pack(pady=5)
result_label = tk.Label(window, text="")
result_label.pack(pady=10)

# Iniciar a aplicação
window.mainloop()
```
