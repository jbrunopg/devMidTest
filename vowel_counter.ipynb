```python
import tkinter as tk

def count_vowels():
    sentence = entry.get()
    vowels = "AEIOUaeiou"
    vowel_count = sum(1 for char in sentence if char in vowels)
    result_text.delete(1.0, tk.END)
    result_text.insert(tk.END, f"The sentence has {vowel_count} vowel(s).")

# Create the main window
root = tk.Tk()
root.title("Vowel Counter")

# Create widgets
label = tk.Label(root, text="Enter a sentence:")
label.pack(pady=10)

entry = tk.Entry(root)
entry.pack(pady=5)

button = tk.Button(root, text="Count Vowels", command=count_vowels)
button.pack(pady=10)

result_text = tk.Text(root, width=30, height=5)
result_text.pack()

# Run the Tkinter event loop
root.mainloop()
```
