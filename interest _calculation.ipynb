```python
import tkinter as tk

def calculate_final_value():
    try:
        principal = float(entry_principal.get())
        interest_rate = float(entry_interest_rate.get()) / 100
        months = int(entry_months.get())

        final_value = principal * (1 + interest_rate * months)
        result_text.config(state=tk.NORMAL)  # Enable text widget for writing
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, f"The final value of the investment is: {final_value:.2f}")
        result_text.config(state=tk.DISABLED)  # Disable text widget for writing

    except ValueError:
        result_text.config(state=tk.NORMAL)  # Enable text widget for writing
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, "Invalid input. Please enter valid numeric values.")
        result_text.config(state=tk.DISABLED)  # Disable text widget for writing

# Create the main window
root = tk.Tk()
root.title("Investment Final Value Calculator")

# Create widgets for investment details
label_principal = tk.Label(root, text="Enter initial capital:")
label_principal.pack(pady=5)
entry_principal = tk.Entry(root)
entry_principal.pack(pady=5)

label_interest_rate = tk.Label(root, text="Enter interest rate (%):")
label_interest_rate.pack(pady=5)
entry_interest_rate = tk.Entry(root)
entry_interest_rate.pack(pady=5)

label_months = tk.Label(root, text="Enter investment time (in months):")
label_months.pack(pady=5)
entry_months = tk.Entry(root)
entry_months.pack(pady=5)

# Create a button to calculate final value
button = tk.Button(root, text="Calculate Final Value", command=calculate_final_value)
button.pack(pady=10)

# Create a text area to display the result
result_text = tk.Text(root, width=30, height=3, state=tk.DISABLED)
result_text.pack()

# Run the Tkinter event loop
root.mainloop()
```
