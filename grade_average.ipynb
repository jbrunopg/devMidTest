```python
import tkinter as tk

def calculate_average():
    try:
        grade1 = float(entry_subject1.get())
        grade2 = float(entry_subject2.get())
        grade3 = float(entry_subject3.get())

        average = (grade1 + grade2 + grade3) / 3
        result_text.config(state=tk.NORMAL)  # Enable text widget for writing
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, f"The average grade is: {average:.2f}")
        result_text.config(state=tk.DISABLED)  # Disable text widget for writing

    except ValueError:
        result_text.config(state=tk.NORMAL)  # Enable text widget for writing
        result_text.delete(1.0, tk.END)  # Clear the result text area
        result_text.insert(tk.END, "Invalid input. Please enter valid numeric grades.")
        result_text.config(state=tk.DISABLED)  # Disable text widget for writing

# Create the main window
root = tk.Tk()
root.title("Grade Average Calculator")

# Create widgets for subject grades
label_subject1 = tk.Label(root, text="Enter grade for subject 1:")
label_subject1.pack(pady=5)
entry_subject1 = tk.Entry(root)
entry_subject1.pack(pady=5)

label_subject2 = tk.Label(root, text="Enter grade for subject 2:")
label_subject2.pack(pady=5)
entry_subject2 = tk.Entry(root)
entry_subject2.pack(pady=5)

label_subject3 = tk.Label(root, text="Enter grade for subject 3:")
label_subject3.pack(pady=5)
entry_subject3 = tk.Entry(root)
entry_subject3.pack(pady=5)

# Create a button to calculate average
button = tk.Button(root, text="Calculate Average", command=calculate_average)
button.pack(pady=10)

# Create a text area to display the result
result_text = tk.Text(root, width=30, height=3, state=tk.DISABLED)
result_text.pack()

# Run the Tkinter event loop
root.mainloop()
```
