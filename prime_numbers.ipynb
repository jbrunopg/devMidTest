```python
import tkinter as tk

def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

def print_first_n_primes(n):
    primes_found = 0
    number = 2
    prime_numbers = []
    while primes_found < n:
        if is_prime(number):
            prime_numbers.append(number)
            primes_found += 1
        number += 1
    return prime_numbers

def check_prime():
    input_number = int(entry.get())
    result = is_prime(input_number)
    if result:
        result_label.config(text=f"{input_number} é um número primo!")
    else:
        result_label.config(text=f"{input_number} não é um número primo!")

def show_first_10_primes():
    prime_numbers = print_first_n_primes(10)
    result_label.config(text=f"Os primeiros 10 números primos são: {', '.join(map(str, prime_numbers))}")

# Configuração básica da janela
window = tk.Tk()
window.title("Verificar Números Primos")
window.geometry("300x200")

# Elementos da interface
entry_label = tk.Label(window, text="Digite um número:")
entry_label.pack(pady=10)
entry = tk.Entry(window)
entry.pack()
check_button = tk.Button(window, text="Verificar Primo", command=check_prime)
check_button.pack(pady=5)
result_label = tk.Label(window, text="")
result_label.pack(pady=10)
show_primes_button = tk.Button(window, text="Mostrar Primeiros 10 Primos", command=show_first_10_primes)
show_primes_button.pack()

# Iniciar a aplicação
window.mainloop()
```
